```python
import numpy as np
```

```python
text = " .txt"
file = open(text,"r")
List = file.read().split("\n")
file.close()
```

```python
def In(lists):
    List=lists.copy()
    length = int(List.pop(0))

    Picture =  {}

    Pic_V = {}
    Pic_H = {}

    for i in range(length):
        Pic = List.pop(0).split(" ",2)
        Picture[i] = [Pic[0],int(Pic[1]),Pic[2].split(" ")]
    
    for i in range(length):
        if Picture[i][0] == "H":
            Pic_H[i]=Picture[i][2]
        elif Picture[i][0] == "V":
            Pic_V[i]=Picture[i][2]
        
    return (length,Picture,Pic_V,Pic_H)
```

```python
length,Picture,Pic_V,Pic_H=In(List)
```

```python
def mis(L_V):
    Out={}
    temp=[]
    for key, value in L_V.items():
        temp.append([key,value])
        if len(temp)==2:
            ele=list(np.union1d(temp[0][1],temp[1][1]))
            Out[temp[0][0],temp[1][0]]=ele
            temp=[]
    return (Out)
```

```python
def chose(slide,L_Slides):
    maxs = 0
    ind=L_Slides[0]
    m=0
    for i in L_Slides:
        leng = min(len(i),len(slide))
        if leng > 1:
            num = len(np.intersect1d(slide,i))
            if num >maxs and num <= leng//2:
                maxs = num
                ind = i
                m=0
        if m==6:
            break
    return (ind)
```

```python
Solution=[]
picture=Picture.copy()
L_Slides=[]
L_sl={}
for key, value in Pic_H.items():
    L_Slides.append(value)
    L_sl[key]=value
    
L_V=mis(Pic_V)
for key, value in L_V.items():
    L_Slides.append(value)
    L_sl[key]=value

slide=L_Slides.pop(0)
ke=list(L_sl.keys())[list(L_sl.values()).index(slide)]
Solution.append(ke)
while L_Slides:
    slide=chose(slide,L_Slides)
    L_Slides.remove(slide)
    ke=list(L_sl.keys())[list(L_sl.values()).index(slide)]
    Solution.append(ke)
    print(Solution[len(Solution)-1])
print(Solution)
```

```python
out=""
out+=str(len(Solution))+'\n'
for i in Solution:
    if type(i)==tuple:
        out+=str(i[0])+" "+str(i[1])+"\n"
    elif type(i)==int :
        out+=str(i)+"\n"
file=open("e.out","w")
file.write(out)
file.close()
```
