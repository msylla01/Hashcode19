```python
import numpy as np
```

```python
text = 'input_8.txt'
file = open(text,'r')
List = file.read().split('\n')
```

```python
def In(List):
    size = List.pop(0)
    size = [int(size.split(',')[0]),int(size.split(',')[1])]
    shape_I = int(List.pop(0))
    
    objects = []
    for i in range(shape_I):
        obj = List.pop(0).split(',')
        objects.append([int(obj[0]),int(obj[1])])
    
    shape_p=int(List.pop(0))
    pairs=[]
    while List:
        temp=List.pop(0).split(",")
        pairs.append([int(temp[0]),int(temp[1])])
    return (size,objects,shape_p,pairs)    
```

```python
size,objects,shape_p,pairs=In(List)
```

```python
friends={}
friendLs=[]
for j in range(shape_p):
    fr=[]
    for k in pairs:
        if j==k[0]:
            fr.append(k[1])
        elif j==k[1]:
            fr.append(k[0])
    friendLs.append([len(fr),j])
    friends[j]=fr
```

```python
friendLs.sort(reverse=True)
```

```python
def Chose(friend,friends):
    init=friend[0]
    for i in friend:
        if i!=init and len(friends[i])>len(friends[init]):
            init=i
    #friend.remove(init)
    return (init)
```

```python
def Arround(i,j):
    if i==0 and j==0:
        return ([[0,1],[1,0]])
    elif i==0 and j==size[0]-1:
        return ([[0,j-1],[1,j]])
    elif i==size[1]-1 and j==0:
        return ([[i-1,j],[i,1]])
    elif i==size[1]-1 and j==size[0]-1:
        return ([[i-1,j],[i,j-1]])
    elif i>0 and i<size[1]-1 and j==0:
        return ([[i-1,0],[i,1],[i+1,0]])
    elif i>0 and i<size[1]-1 and j==size[0]-1:
        return ([[i-1,j],[i,j-1],[i+1,j]])
    elif j>0 and j<size[0]-1 and i==0:
        return ([[0,j-1],[1,j],[0,j+1]])
    elif j>0 and j<size[0]-1 and i==size[1]-1:
        return ([[i,j-1],[i,j+1],[i-1,j]])
    else:
        return ([[i-1,j],[i,j+1],[i,j-1],[i+1,j]])
```

```python
def position(p):
    for i in range(len(Block)):
        for j in range(len(Block[0])):
             if Block[i,j]==p:
                return [i,j]
```

```python
def place(lists,friends):
    for i in lists:
        m,n=position(i)
        ar=Arround(m,n)
        for k in ar:
            if Block[k[0],k[1]]==-1:
                return k
    return None
```

```python
def lis(p,friends):
    fr=friends[p]
    l=[]
    for row in Block:
        for col in row:
            if col in fr:
                l.append(col)
    return l
```

```python
def put(i,j,p,Block):
    Block[i][j]=p
    Solution.append([p,j,i])
    t=len(Solution)
    print(str(t)+"    "+str(Solution[t-1]))
```
